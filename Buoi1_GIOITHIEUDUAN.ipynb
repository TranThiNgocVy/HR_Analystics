## 🌏 Business Context: Tình hình hiện nay

### **"War for Talent" ở Việt Nam 2026**

**Sự thực hiện nay:**
- 🔄 Cạnh tranh "chiều dọc - chiều ngang": các startup AI, blockchain, SaaS tuyển dụng "tổn hao" nhân tài từ tập đoàn lớn
- 📈 **Recruitment cost** tăng 30-40% (signing bonus, flexible benefit)
- ⚡ Nhân sự tech chuyển việc nhanh do:
  - Job hopping để tăng lương 20-30%
  - Thiếu career path rõ ràng
  - Burnout (overtime culture)
  - Remote/hybrid mismatch

### **Tác động kinh tế**
- 🔴 Chi phí thay thế 1 nhân viên giỏi = **150% salary** (tuyển, training, onboarding)
- 💰 Giảm 10% turnover = tiết kiệm **200-300M VNĐ/năm** cho công ty 500 người
- ⏰ Thời gian đào tạo mất ~3 tháng, năng suất 60% → 100%

1. 📊 Phát hiện sớm dấu hiệu churn (engagement ↓, performance chệch)
2. ⚖️ Xác định lương công bằng (gender gap, level gap)
3. 🎯 Đề xuất retention program kịp thời

## ⚙️ Cài đặt PostgreSQL + kết nối + query tool

### Download và cài PostgreSQL
1. Truy cập: https://www.postgresql.org/download/
2. Chọn hệ điều hành (Windows/macOS/Linux) → tải installer.
3. Chạy installer:
   - PostgreSQL server
   - pgAdmin (GUI)
   - (Stack Builder tuỳ chọn)
4. Ghi nhớ password `postgres`, port mặc định 5432.

### Khởi động PostgreSQL và tạo database
1. Mở `pgAdmin 4` hoặc `psql` terminal.
2. Kết nối server `localhost:5432` bằng user `postgres`.
3. Tạo database:
   ```sql
   CREATE DATABASE hr_analytics;
   ```
4. Trong pgAdmin: chuột phải `Databases` → `Create` → `Database...` nếu dùng GUI.

### Mở Query Tool
1. Trong pgAdmin: chọn database `hr_analytics` phải chuột chọn → _Query Tool_.
2. DBeaver/Datagrip: tạo connection đến PostgreSQL, chọn database `hr_analytics`, mở SQL editor.

## ⚙️ BƯỚC 1: Tạo bảng (CREATE TABLE)

**Dán lệnh SQL dưới đây vào PostgreSQL client của bạn:**<br>
<span style="color:red">*(Tô đen đoạn lệnh - Nhấn F5 hoặc nút Excute Script trên cửa sổ Query Tool)*<span>

```sql
-- Drop nếu đã tồn tại
DROP TABLE IF EXISTS leavers CASCADE;
DROP TABLE IF EXISTS engagement_surveys CASCADE;
DROP TABLE IF EXISTS performance_reviews CASCADE;
DROP TABLE IF EXISTS salaries CASCADE;
DROP TABLE IF EXISTS employees CASCADE;
DROP TABLE IF EXISTS job_titles CASCADE;
DROP TABLE IF EXISTS departments CASCADE;

-- 1. DEPARTMENTS
CREATE TABLE departments (
  dept_id SERIAL PRIMARY KEY,
  dept_name TEXT NOT NULL,
  location TEXT NOT NULL,
  head_id INT NULL
);

-- 2. JOB_TITLES
CREATE TABLE job_titles (
  title_id SERIAL PRIMARY KEY,
  title_name TEXT NOT NULL,
  level TEXT CHECK(level IN ('Junior','Mid','Senior','Lead','Manager','Director')),
  track TEXT CHECK(track IN ('Engineering','Product','Design','HR','Finance','Sales'))
);

-- 3. EMPLOYEES
CREATE TABLE employees (
  emp_id SERIAL PRIMARY KEY,
  first_name TEXT NOT NULL,
  last_name TEXT NOT NULL,
  gender CHAR(1) CHECK(gender IN ('M','F','O')),
  date_of_birth DATE NOT NULL,
  hired_date DATE NOT NULL,
  dept_id INT REFERENCES departments(dept_id),
  title_id INT REFERENCES job_titles(title_id),
  manager_id INT REFERENCES employees(emp_id),
  employment_status TEXT CHECK(employment_status IN ('active','on_leave','terminated')),
  city TEXT NOT NULL,
  remote_status TEXT CHECK(remote_status IN ('Remote','Hybrid','Onsite')),
  education_level TEXT
);

-- 4. SALARIES
CREATE TABLE salaries (
  salary_id SERIAL PRIMARY KEY,
  emp_id INT REFERENCES employees(emp_id),
  base_salary NUMERIC(12,2) NOT NULL,
  bonus NUMERIC(12,2),
  effective_date DATE NOT NULL
);

-- 5. PERFORMANCE_REVIEWS
CREATE TABLE performance_reviews (
  review_id SERIAL PRIMARY KEY,
  emp_id INT REFERENCES employees(emp_id),
  review_date DATE NOT NULL,
  performance_score INT CHECK(performance_score BETWEEN 1 AND 10),
  rating TEXT CHECK(rating IN ('Below expectation','Meets expectation','Exceeds expectation')),
  promo_recommendation BOOLEAN
);

-- 6. ENGAGEMENT_SURVEYS
CREATE TABLE engagement_surveys (
  response_id SERIAL PRIMARY KEY,
  emp_id INT REFERENCES employees(emp_id),
  survey_date DATE NOT NULL,
  engagement_score INT CHECK(engagement_score BETWEEN 1 AND 10),
  wellbeing_score INT CHECK(wellbeing_score BETWEEN 1 AND 10),
  nps_score INT CHECK(nps_score BETWEEN -100 AND 100)
);

-- 7. LEAVERS
CREATE TABLE leavers (
  leaver_id SERIAL PRIMARY KEY,
  emp_id INT REFERENCES employees(emp_id),
  leaving_date DATE NOT NULL,
  reason TEXT,
  exit_type TEXT CHECK(exit_type IN ('Voluntary','Involuntary','Retirement')),
  rehired BOOLEAN DEFAULT FALSE
);

-- Create indexes
CREATE INDEX idx_employees_dept ON employees(dept_id);
CREATE INDEX idx_employees_status ON employees(employment_status);
CREATE INDEX idx_salaries_emp ON salaries(emp_id);
CREATE INDEX idx_performance_emp ON performance_reviews(emp_id);
CREATE INDEX idx_engagement_emp ON engagement_surveys(emp_id);
```

```sql
SELECT 
  d.dept_name,
  COUNT(e.emp_id) as emp_count
FROM employees e
LEFT JOIN departments d ON e.dept_id = d.dept_id
GROUP BY d.dept_name
ORDER BY emp_count DESC;
```
