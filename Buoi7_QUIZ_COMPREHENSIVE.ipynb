## ❓ Câu 1 (Cơ bản): SELECT & WHERE

**Câu hỏi:** Để lấy tên và lương của nhân viên ở phòng Engineering (dept_id = 1), query nào đúng?

- A) `SELECT name, salary FROM employees WHERE dept_id = 1`
- B) `SELECT first_name, last_name, base_salary FROM employees WHERE dept_id = 1` ✅
- C) `SELECT * FROM employees WHERE dept_id != 1`
- D) `SELECT employees FROM employees WHERE dept_id = 1`

**Đáp án:** B

**Giải thích:** Cột lương là `base_salary` (không phải `salary`). `first_name` và `last_name` là tên chính xác (không phải `name`). Điều kiện `WHERE dept_id = 1` lọc phòng Engineering.

## ❓ Câu 2 (Cơ bản): COUNT & GROUP BY

**Câu hỏi:** Để đếm số nhân viên theo từng phòng ban, query nào đúng?

- A) `SELECT dept_id, COUNT(*) FROM employees`
- B) `SELECT dept_id, COUNT(*) FROM employees GROUP BY dept_id` ✅
- C) `SELECT dept_id, COUNT(emp_id) FROM employees WHERE dept_id > 0`
- D) `SELECT COUNT(*) FROM employees GROUP BY dept_id`

**Đáp án:** B

**Giải thích:** `GROUP BY dept_id` là bắt buộc khi dùng hàm tổng hợp COUNT. Câu A thiếu GROUP BY. Câu C thiếu GROUP BY. Câu D lập lại nhưng thiếu `dept_id` trong SELECT (sẽ gây lỗi SQL).

## ❓ Câu 3 (Cơ bản): JOIN & INNER JOIN

**Câu hỏi:** Để lấy tên nhân viên + tên phòng ban, cần đặt điều kiện JOIN như thế nào?

- A) `FROM employees e JOIN departments d ON e.emp_id = d.dept_id`
- B) `FROM employees e JOIN departments d ON e.dept_id = d.dept_id` ✅
- C) `FROM employees e JOIN departments d ON e.first_name = d.dept_name`
- D) `FROM employees e, departments d WHERE e.dept_id = d.dept_id`

**Đáp án:** B

**Giải thích:** Khóa ngoại `dept_id` trong bảng employees liên kết với khóa chính `dept_id` trong departments. Câu A sai (khớp sai cột khóa), Câu C sai (khớp sai kiểu dữ liệu), Câu D là cú pháp cũ (comma join) nhưng chứng thực đúng dữ liệu (nhưng B là chuẩn mực INNER JOIN).

## ❓ Câu 4 (Trung cấp): DATE Functions & EXTRACT

**Câu hỏi:** Để tính số năm kinh nghiệm của nhân viên từ ngày tuyển dụng (`hired_date`) đến hôm nay, query nào đúng?

- A) `SELECT first_name, EXTRACT(YEAR FROM hired_date) FROM employees`
- B) `SELECT first_name, EXTRACT(YEAR FROM AGE(NOW(), hired_date)) FROM employees` ✅
- C) `SELECT first_name, AGE(hired_date) FROM employees`
- D) `SELECT first_name, YEAR(NOW()) - YEAR(hired_date) FROM employees`

**Đáp án:** B

**Giải thích:** Hàm `AGE(NOW(), hired_date)` tính khoảng thời gian (interval). `EXTRACT(YEAR FROM ...)` lấy số năm từ interval đó. Câu A chỉ lấy năm của ngày tuyển dụng. Câu C trả về interval (không phải số). Câu D là cú pháp MySQL, PostgreSQL không hỗ trợ YEAR().

## ❓ Câu 5 (Trung cấp): CASE WHEN

**Câu hỏi:** Để phân loại nhân viên theo mức độ "New / Senior / Veteran" dựa trên năm tuyển dụng, query nào đúng?

```sql
SELECT 
  first_name,
  CASE 
    WHEN EXTRACT(YEAR FROM AGE(NOW(), hired_date)) < 2 THEN 'New'
    WHEN EXTRACT(YEAR FROM AGE(NOW(), hired_date)) < 5 THEN 'Senior'
    ELSE 'Veteran'
  END as experience_level
FROM employees
```

**Câu hỏi thực:** Kết quả của query trên là gì khi nhân viên A được tuyển năm 2020?

- A) 'New' (vì < 2 năm)
- B) 'Senior' (vì 2-5 năm) ✅
- C) 'Veteran' (vì > 5 năm)
- D) NULL (lỗi syntax)

**Đáp án:** B

**Giải thích:** 2020 → 2026 = 6 năm kinh nghiệm. Điều kiện đầu (< 2) sai. Điều kiện thứ hai (< 5) sai. Nên rơi vào ELSE → 'Veteran'. *Lưu ý: Nếu năm 2025 thì là 'Senior'. Đáp án giả định hiện tại là 2026.*

## ❓ Câu 6 (Trung cấp): Churn Rate Calculation

**Câu hỏi:** Để tính churn rate (tỷ lệ rời bỏ) trong năm 2025 = (số leavers / tổng đầu năm) × 100, query nào thích hợp?

- A) `SELECT COUNT(DISTINCT emp_id) FROM leavers WHERE EXTRACT(YEAR FROM leave_date) = 2025`
- B) `SELECT ROUND((COUNT(DISTINCT l.emp_id) / COUNT(DISTINCT e.emp_id)) * 100, 2) FROM leavers l CROSS JOIN employees e WHERE EXTRACT(YEAR FROM l.leave_date) = 2025` ✅
- C) `SELECT COUNT(*) FROM employees WHERE hired_date = '2025-01-01'`
- D) `SELECT SUM(leave_date) FROM leavers WHERE EXTRACT(YEAR FROM leave_date) = 2025`

**Đáp án:** B

**Giải thích:** Churn rate = (số người rời năm 2025 / tổng số nhân viên hiện tại) × 100. Cách B dùng CROSS JOIN để lấy tổng employees. ROUND(..., 2) làm tròn 2 chữ số thập phân. Các câu khác là đếm/tính từng phần hoặc sai logic.

## ❓ Câu 7 (Nâng cao): Window Functions - RANK vs ROW_NUMBER

**Câu hỏi:** Để xếp hạng nhân viên theo lương (cao nhất trước) trong mỗi phòng ban, và xử lý nếu 2 nhân viên cùng lương:

```sql
SELECT 
  dept_id,
  emp_id,
  base_salary,
  RANK() OVER (PARTITION BY dept_id ORDER BY base_salary DESC) as rank_salary
FROM salaries
```

**Câu hỏi thực:** Nếu phòng A có 4 nhân viên với lương (50M, 45M, 45M, 40M), rank sẽ là?

- A) 1, 2, 3, 4 (ROW_NUMBER)
- B) 1, 2, 2, 4 (RANK) ✅
- C) 1, 2, 2, 3 (DENSE_RANK)
- D) 1, 1, 1, 1 (GROUP BY)

**Đáp án:** B

**Giải thích:** `RANK()` cho cùng hạng khi bằng nhau, rồi nhảy cóc (rank 2 người → rank 4 tiếp theo). `ROW_NUMBER()` không bao giờ trùng. `DENSE_RANK()` không nhảy cóc (2 người rank 2 → rank 3 tiếp theo).

## ❓ Câu 8 (Nâng cao): LAG Window Function & Salary History

**Câu hỏi:** Để tính mức tăng lương (current - previous) của mỗi nhân viên theo thời gian, query nào đúng?

```sql
SELECT 
  emp_id,
  effective_date,
  base_salary,
  LAG(base_salary) OVER (PARTITION BY emp_id ORDER BY effective_date) as prev_salary,
  base_salary - LAG(base_salary) OVER (PARTITION BY emp_id ORDER BY effective_date) as salary_increase
FROM salaries
ORDER BY emp_id, effective_date
```

**Câu hỏi thực:** Kết quả của nhân viên có salary history (2023: 30M, 2024: 33M, 2025: 36M) là?

- A) salary_increase = (NULL, 3M, 3M) ✅
- B) salary_increase = (0, 3M, 3M)
- C) salary_increase = (30M, 33M, 36M)
- D) salary_increase = Error (không thể trừ NULL)

**Đáp án:** A

**Giải thích:** `LAG()` hàng đầu trả về NULL (không có hàng trước). Ở hàng 2: LAG = 30M, 33M - 30M = 3M. Ở hàng 3: LAG = 33M, 36M - 33M = 3M.

## ❓ Câu 9 (Nâng cao): CTEs (Common Table Expressions) & Subqueries

**Câu hỏi:** Để tìm phòng ban nào có lương trung bình cao nhất, query nào sử dụng CTE chính xác?

- A) `WITH avg_salary AS (SELECT dept_id, AVG(base_salary) as avg_sal FROM salaries GROUP BY dept_id) SELECT * FROM avg_salary ORDER BY avg_sal DESC LIMIT 1` ✅
- B) `SELECT dept_id, AVG(base_salary) as max_avg FROM salaries GROUP BY dept_id ORDER BY max_avg DESC LIMIT 1`
- C) `WITH dept_salaries AS (SELECT * FROM salaries) SELECT dept_id FROM dept_salaries`
- D) `SELECT TOP 1 dept_id, AVG(base_salary) FROM salaries ORDER BY AVG(base_salary) DESC`

**Đáp án:** A

**Giải thích:** Câu A dùng CTE `avg_salary` để tính AVG lương theo dept_id, rồi ORDER BY DESC, LIMIT 1 để lấy tối đa. Câu B không dùng CTE (nhưng cũng đúng kết quả). Câu C CTE nhưng không tính gì. Câu D dùng TOP (T-SQL), PostgreSQL dùng LIMIT.

## ❓ Câu 10 (Nâng cao): Burnout Detection & Multiple Conditions

**Câu hỏi:** Để phát hiện "burnout risk" (nhân viên có engagement score < 50 VÀ performance score > 80), query nào đúng?

```sql
SELECT 
  e.emp_id,
  e.first_name,
  es.engagement_score,
  pr.performance_score,
  CASE 
    WHEN es.engagement_score < 50 AND pr.performance_score > 80 THEN 'Burnout Risk'
    ELSE 'Normal'
  END as status
FROM employees e
JOIN engagement_surveys es ON e.emp_id = es.emp_id
JOIN performance_reviews pr ON e.emp_id = pr.emp_id
WHERE es.engagement_score < 50 AND pr.performance_score > 80
```

**Câu hỏi thực:** Query này có vấn đề gì không?

- A) Không vấn đề, query hoàn toàn chính xác
- B) Lệnh WHERE là dư thừa (vì CASE WHEN đã lọc), tuy nhiên query vẫn chạy đúng ✅
- C) Phần JOIN sai, nên dùng LEFT JOIN
- D) Không thể JOIN 3 bảng cùng lúc

**Đáp án:** B

**Giải thích:** Query không có lỗi cú pháp, nhưng WHERE clause là dư thừa vì kết quả status đã được CASE WHEN xử lý. Nếu muốn chỉ lấy burnout risk, nên loại bỏ WHERE và lọc WHERE status = 'Burnout Risk', hoặc loại bỏ CASE WHEN và chỉ dùng WHERE.
