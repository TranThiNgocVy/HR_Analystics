## 🌏 Đề tài Employee Overview Analysis là gì?

**Employee Overview Analysis** là phân tích tổng quan về cơ cấu nhân sự trong tổ chức. Đây là nền tảng của HR Analytics, giúp doanh nghiệp:

- 👥 **Hiểu cơ cấu**: Nhân viên phân bố như thế nào?
- 📊 **Phân tích nhân khẩu học**: Tuổi, giới tính, vị trí
- 🏢 **Xem xét tổ chức**: Số lượng theo phòng ban, cấp bậc
- 🎯 **Đề xuất chiến lược**: Tuyển dụng, đào tạo phù hợp

### 📊 Ví dụ điển hình ở Việt Nam (2026)
- **Doanh nghiệp lớn**: 70% nhân viên dưới 35 tuổi, tỷ lệ nữ 40%
- **Startup Tech**: Trung bình 28 tuổi, nam chiếm 65%
- **Ngành Tài chính**: Tỷ lệ nữ cao hơn nam, đặc biệt ở vị trí quản lý

**Giá trị kinh doanh**: Giúp tối ưu hóa chi phí tuyển dụng, giảm turnover bằng cách hiểu rõ cơ cấu nhân sự.

### ❓ Các truy vấn thường được hỏi
- Tổng số nhân viên theo phòng ban
- Phân bố nhân viên theo độ tuổi
- Tỷ lệ nam/nữ trong công ty
- Số lượng nhân viên theo vị trí công việc

## 🔍 BƯỚC 1: SELECT Cơ bản - Chọn dữ liệu từ bảng

**SELECT** là câu lệnh đầu tiên bạn học trong SQL. Nó dùng để chọn dữ liệu từ bảng.

### 1.1 SELECT tất cả cột (*)
```sql
SELECT * FROM employees LIMIT 5;
```
**Kỳ vọng:** Hiển thị 5 dòng đầu tiên với tất cả cột của bảng employees.

![img3-1.png](images/images_buoi3/img3-1.png)

### 1.2 SELECT cột cụ thể
```sql
SELECT first_name, last_name, gender FROM employees LIMIT 5;
```
**Kỳ vọng:** Chỉ hiển thị 3 cột: first_name, last_name, gender.

![img3-2.png](images/images_buoi3/img3-2.png)

### 1.3 SELECT với alias (AS)
```sql
SELECT 
  first_name || ' ' || last_name AS full_name,
  gender,
  hired_date
FROM employees 
LIMIT 5;
```
**Kỳ vọng:** Cột đầu tiên có tên "full_name" là kết hợp first_name + last_name.

![img3-3.png](images/images_buoi3/img3-3.png)

**Giải thích:** `||` là toán tử nối chuỗi trong PostgreSQL. `AS` đặt tên bí danh cho cột.

## 🔍 BƯỚC 2: WHERE - Lọc dữ liệu theo điều kiện

**WHERE** dùng để lọc dữ liệu theo điều kiện cụ thể.

### 2.1 WHERE với điều kiện bằng (=)
```sql
SELECT first_name, last_name, dept_id 
FROM employees 
WHERE dept_id = 1 
LIMIT 5;
```
**Kỳ vọng:** Chỉ nhân viên phòng ban ID = 1 (Engineering).

![img3-4.png](images/images_buoi3/img3-4.png)

### 2.2 WHERE với nhiều điều kiện (AND, OR)
```sql
SELECT first_name, last_name, gender, city
FROM employees 
WHERE gender = 'F' AND city = 'Hanoi'
LIMIT 5;
```
**Kỳ vọng:** Nữ nhân viên ở Hà Nội.

![img3-5.png](images/images_buoi3/img3-5.png)

### 2.3 WHERE với so sánh (>, <, >=, <=)
```sql
SELECT first_name, last_name, hired_date
FROM employees 
WHERE hired_date >= '2023-01-01'
LIMIT 5;
```
**Kỳ vọng:** Nhân viên tuyển dụng từ 2023 trở đi.

![img3-6.png](images/images_buoi3/img3-6.png)

**Giải thích:** `AND` yêu cầu cả hai điều kiện đúng. `OR` chỉ cần một trong hai đúng.

## 🔍 BƯỚC 3: ORDER BY - Sắp xếp dữ liệu

**ORDER BY** sắp xếp kết quả theo cột chỉ định.

### 3.1 ORDER BY tăng dần (ASC)
```sql
SELECT first_name, last_name, hired_date
FROM employees 
ORDER BY hired_date ASC
LIMIT 5;
```
**Kỳ vọng:** Nhân viên cũ nhất trước (hired_date sớm nhất).

![img3-7.png](images/images_buoi3/img3-7.png)

### 3.2 ORDER BY giảm dần (DESC)
```sql
SELECT first_name, last_name, hired_date
FROM employees 
ORDER BY hired_date DESC
LIMIT 5;
```
**Kỳ vọng:** Nhân viên mới nhất trước (hired_date muộn nhất).

![img3-8.png](images/images_buoi3/img3-8.png)

### 3.3 ORDER BY nhiều cột
```sql
SELECT dept_id, first_name, last_name, hired_date
FROM employees 
ORDER BY dept_id ASC, hired_date DESC
LIMIT 10;
```
**Kỳ vọng:** Sắp xếp theo dept_id tăng dần, trong mỗi dept thì hired_date giảm dần.

![img3-9.png](images/images_buoi3/img3-9.png)

**Giải thích:** `ASC` là mặc định (tăng dần). `DESC` là giảm dần. Có thể sắp xếp theo nhiều cột.

## 🔍 BƯỚC 4: JOIN - Kết hợp dữ liệu từ nhiều bảng

**JOIN** kết hợp dữ liệu từ 2+ bảng dựa trên cột chung (thường là khóa ngoại).

### 4.1 INNER JOIN - Chỉ lấy dữ liệu có ở cả hai bảng
```sql
SELECT 
  e.first_name,
  e.last_name,
  d.dept_name
FROM employees e
INNER JOIN departments d ON e.dept_id = d.dept_id
LIMIT 5;
```
**Kỳ vọng:** Hiển thị tên nhân viên + tên phòng ban.

![img3-10.png](images/images_buoi3/img3-10.png)

### 4.2 LEFT JOIN - Lấy tất cả từ bảng trái, null nếu không match
```sql
SELECT 
  e.first_name,
  e.last_name,
  d.dept_name
FROM employees e
LEFT JOIN departments d ON e.dept_id = d.dept_id
WHERE d.dept_name IS NULL
LIMIT 5;
```
**Kỳ vọng:** Nhân viên không có phòng ban (dept_id null hoặc không tồn tại) -> Bảng rỗng vì tất cả nhân viên được khởi tạo ban đầu đều có phòng ban.

![img3-11.png](images/images_buoi3/img3-11.png)

**Giải thích:** `INNER JOIN` chỉ lấy records match. `LEFT JOIN` giữ tất cả từ bảng trái (employees), null nếu không match ở bảng phải (departments).

## 🔍 BƯỚC 5: GROUP BY & Aggregate Functions

**GROUP BY** nhóm dữ liệu theo cột, kết hợp với hàm tổng hợp.

### 5.1 COUNT - Đếm số lượng
```sql
SELECT 
  dept_id,
  COUNT(*) as emp_count
FROM employees
GROUP BY dept_id
ORDER BY emp_count DESC;
```
**Kỳ vọng:** Số nhân viên theo từng dept_id.

![img3-12.png](images/images_buoi3/img3-12.png)

### 5.2 SUM - Tổng cộng
```sql
SELECT 
  dept_id,
  SUM(base_salary) as total_salary
FROM employees e
JOIN salaries s ON e.emp_id = s.emp_id
GROUP BY dept_id
ORDER BY total_salary DESC;
```
**Kỳ vọng:** Tổng lương theo phòng ban.

![img3-13.png](images/images_buoi3/img3-13.png)

### 5.3 AVG - Trung bình
```sql
SELECT 
  dept_id,
  ROUND(AVG(base_salary), 0) as avg_salary
FROM employees e
JOIN salaries s ON e.emp_id = s.emp_id
GROUP BY dept_id
ORDER BY avg_salary DESC;
```
**Kỳ vọng:** Lương trung bình theo phòng ban (làm tròn).

![img3-14.png](images/images_buoi3/img3-14.png)

**Giải thích:** `COUNT(*)` đếm tất cả rows. `SUM(column)` tổng giá trị. `AVG(column)` trung bình. `ROUND(value, decimals)` làm tròn số.

## 💪 BƯỚC 6: 5 Bài tập đơn giản (Practice Exercises)

### Bài 1: Liệt kê tất cả nhân viên nữ
**Yêu cầu:** Viết query hiển thị first_name, last_name của nhân viên nữ, sắp xếp theo tên.

**Đáp án:**
```sql
SELECT first_name, last_name
FROM employees
WHERE gender = 'F'
ORDER BY first_name;
```

![img3-15.png](images/images_buoi3/img3-15.png)

**Giải thích:** `WHERE gender = 'F'` lọc nữ. `ORDER BY first_name` sắp xếp A-Z.

---

### Bài 2: Đếm số nhân viên theo thành phố
**Yêu cầu:** Viết query đếm số nhân viên ở mỗi city.

**Đáp án:**
```sql
SELECT 
  city,
  COUNT(*) as emp_count
FROM employees
GROUP BY city
ORDER BY emp_count DESC;
```

![img3-16.png](images/images_buoi3/img3-16.png)

**Giải thích:** `GROUP BY city` nhóm theo thành phố. `COUNT(*)` đếm số nhân viên mỗi nhóm.

---

### Bài 3: Tìm nhân viên lương cao nhất
**Yêu cầu:** Viết query tìm 5 nhân viên có lương cao nhất.

**Đáp án:**
```sql
SELECT 
  e.first_name,
  e.last_name,
  s.base_salary
FROM employees e
JOIN salaries s ON e.emp_id = s.emp_id
ORDER BY s.base_salary DESC
LIMIT 5;
```

![img3-17.png](images/images_buoi3/img3-17.png)

**Giải thích:** `JOIN salaries` lấy lương. `ORDER BY base_salary DESC` sắp xếp giảm dần. `LIMIT 5` lấy 5 đầu.

---

### Bài 4: Lọc nhân viên theo kinh nghiệm
**Yêu cầu:** Viết query nhân viên tuyển dụng trước 2022.

**Đáp án:**
```sql
SELECT first_name, last_name, hired_date
FROM employees
WHERE hired_date < '2022-01-01'
ORDER BY hired_date DESC;
```

![img3-18.png](images/images_buoi3/img3-18.png)

**Giải thích:** `WHERE hired_date < '2022-01-01'` lọc trước 2022. `ORDER BY hired_date DESC` mới nhất trước.

---

### Bài 5: Thống kê theo giới tính
**Yêu cầu:** Viết query đếm và lương TB theo gender.

**Đáp án:**
```sql
SELECT 
  gender,
  COUNT(*) as count,
  ROUND(AVG(s.base_salary), 0) as avg_salary
FROM employees e
JOIN salaries s ON e.emp_id = s.emp_id
GROUP BY gender;
```

![img3-19.png](images/images_buoi3/img3-19.png)

**Giải thích:** `GROUP BY gender` nhóm theo giới. `COUNT(*)` đếm. `AVG(base_salary)` lương TB.

## 🚀 BƯỚC 7: 4 Bài tập nâng cao (Bonus Exercises)

### Bài 6: Nhân viên và phòng ban
**Yêu cầu:** Viết query hiển thị tên NV + tên phòng + tên job title.

**Đáp án:**
```sql
SELECT 
  e.first_name || ' ' || e.last_name AS full_name,
  d.dept_name,
  jt.title_name
FROM employees e
LEFT JOIN departments d ON e.dept_id = d.dept_id
LEFT JOIN job_titles jt ON e.title_id = jt.title_id
LIMIT 10;
```

**Giải thích:** `LEFT JOIN` giữ tất cả employees. Kết hợp 3 bảng qua FK.

![img3-20.png](images/images_buoi3/img3-20.png)

### Bài 7: Top phòng ban theo số nhân viên
**Yêu cầu:** Viết query top 3 phòng có nhiều nhân viên nhất.

**Đáp án:**
```sql
SELECT 
  d.dept_name,
  COUNT(e.emp_id) as emp_count
FROM employees e
JOIN departments d ON e.dept_id = d.dept_id
GROUP BY d.dept_name
ORDER BY emp_count DESC
LIMIT 3;
```

**Giải thích:** Nhóm theo dept_name, đếm số nhân viên, sắp xếp DESC, LIMIT 3.

![img3-21.png](images/images_buoi3/img3-21.png)

### Bài 8: Nhân viên không có manager
**Yêu cầu:** Viết query nhân viên chưa có quản lý trực tiếp.

**Đáp án:**
```sql
SELECT 
  first_name,
  last_name,
  hired_date
FROM employees
WHERE manager_id IS NULL;
```

**Giải thích:** `WHERE manager_id IS NULL` lọc nhân viên cấp cao nhất không có manager.

![img3-22.png](images/images_buoi3/img3-22.png)

### Bài 9: Thống kê theo thành phố
**Yêu cầu:** Viết query đếm NV theo city và remote_status.

**Đáp án:**
```sql
SELECT 
  city,
  remote_status,
  COUNT(emp_id) as emp_count
FROM employees
GROUP BY city, remote_status
ORDER BY city, emp_count DESC;
```

**Giải thích:** GROUP BY city và remote_status, đếm nhân viên theo từng nhóm.
